## Засвоєння поняття тензору, знати різні способи створення тензорів

```python
import tensorflow as tf
import numpy as np
import pandas as pd

# створення
tensor = tf.Variable([
                   [7, 10, 13, 15],
                   [8, 11, 16, 17],
                   [9, 10, 12, 13],
                  #  [3, 7, 11, 15]
])

print(tf.constant([
                   [7, 10, 13],
                   [8, 12, 16]
]))

print(tf.random.Generator.from_seed(42).normal(shape=(2,3)))
```

## Cтворити набір даних із заданим розподілом (розподіл та розмір за власним рішенням), визначити його властивості (shape, rank, size, axis)

```python
tensor = tf.Variable([
                   [7, 10, 13, 15],
                   [8, 11, 16, 17],
                   [9, 10, 12, 13],
                  #  [3, 7, 11, 15]
])

# визначення властивостей
print(tensor.shape)
print(tf.rank(tensor))
print(tf.size(tensor))
print(tf.expand_dims(tensor, axis=-1))
```

## Продемонструвати основні математичні операції над тензорами

```python
tensor = tf.Variable([
                   [7, 10, 13, 15],
                   [8, 11, 16, 17],
                   [9, 10, 12, 13],
                  #  [3, 7, 11, 15]
])

# основні математичні операції
print(tensor+10)
print(tensor*10)
print(tensor@tf.reshape(tensor, shape=(4, 3)))
```

## Провести CRUD операції над даними, reshape, інтеграцію з np.array, pd.DataFrame)

```python
tensor = tf.Variable([
                   [7, 10, 13, 15],
                   [8, 11, 16, 17],
                   [9, 10, 12, 13],
                  #  [3, 7, 11, 15]
])

# CRUD операції над даними, reshape, інтеграція з np.array, pd.DataFrame
tf.compat.v1.assign(tensor[2], [3, 7, 11, 15])
print(tf.slice(tensor, [1, 0], [1, 4]))
print(tf.reshape(tensor, shape=(4, 3)))
print(tf.Variable(np.array([[2, 4], [6, 9]])))
print(tf.Variable(pd.DataFrame([[2, 4], [6, 9]])))
```
