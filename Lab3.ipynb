#Задача класифікації

З використанням бібліотеки Sklearn вирішити задачу класифікації. 

•	Поглиблено вивчити функцію sklearn.datasets.make_classification

•	Згенерувати набір даних за допомогою означеної функції, для бінарної класифікації. 

•	Здійснити моделювання згенерованого набору даних та візуалізацію результатів.

•	Надати значення метрик якості, приклад класифікації


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import balanced_accuracy_score, f1_score, log_loss, precision_score

# генерація даних
X, y_true = make_classification(n_samples=1000, n_features=5, n_informative=3, n_redundant=0)

# моделювання даних
model = RidgeClassifier().fit(X, y_true)
y_pred = model.predict(X)
print(model.score(X, y_true))

# метрики якості
# обчислення збалансованої точності
print(balanced_accuracy_score(y_true, y_pred))
# обчислення збалансованого показника F
print(f1_score(y_true, y_pred, average='binary'))
# обчислення втрат логістики
print(log_loss(y_true, y_pred))
# обчислення точності
print(precision_score(y_true, y_pred, average='binary'))
```
